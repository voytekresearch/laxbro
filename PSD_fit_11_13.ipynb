```python
%config InlineBackend.figure_format = 'retina'
%pylab inline
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
import scipy.io
import scipy.signal
```

```python
filename = 'gdat.mat'; #filename
filename = os.path.join('/Users/Torben/Desktop/Voytek/MATLAB/', filename) #path to file
data = sp.io.loadmat(filename)
srate = 1000 # sampling rate
data = data['gdat'] # 84 channels of ECoG data
data = data[50, :] #one channel of ECoG data (can be between 1 and 84)
data =  squeeze(data); 
```

```python
#resampling data and calculating PSD using Welch's method

newsrate = 1024 # new sampling rate

if srate!=newsrate: # resample
    data = sp.signal.resample(data, int(np.floor(len(data)*(newsrate/srate)))) 

f,pspec = sp.signal.welch(data, fs=newsrate, window='hanning', nperseg=2*newsrate, noverlap=newsrate/2, nfft=None, detrend='linear', return_onesided=True, scaling='density')
#using Welch's method of calculating PSD. Basically it does a fft on the signal using a sliding window and takes the mean of each frequency. This cleans up the fft a lot.
fwin = [0, 150] 
# this is the  range of frequencies that we're interested in.
#any frequencies with >150 Hz are going to be noise.

fwin = [np.argmin(abs(f-fwin[0])), np.argmin(abs(f-fwin[1]))] # pspec over freq range of interest
f = f[fwin[0]:fwin[1]]
#f is just a variable for the frequency range of interest in 0.5 Hz increments 

#plotting the PSD
pspec = pspec[fwin[0]:fwin[1]]
pspec = np.log10(pspec)
plt.plot(f,pspec)
```

```python
#Removing Troughs
der = (diff(pspec));
mi1 = np.argmin((diff(pspec[100:140])));
mi1 = mi1+100-5

ma1 = np.argmax((diff(pspec[110:150])));
ma1 = ma1+110+5

mi2 = np.argmin((diff(pspec[220:260])));
mi2 = mi2+220-5

ma2 = np.argmax((diff(pspec[230:270])));
ma2 = ma2+230+5

#print find(diff(pspec == mi))
#plt.plot(f[:-1],der)

#values that bridge the troughs
bridge1 = linspace(pspec[mi1],pspec[ma1],ma1-mi1);
bridge2 = linspace(pspec[mi2],pspec[ma2],ma2-mi2);

newpspec = np.concatenate([pspec[:mi1],bridge1])
newpspec = np.concatenate([newpspec,pspec[ma1:]])

newpspec = np.concatenate([newpspec[:mi2],bridge2])
newpspec = np.concatenate([newpspec,pspec[ma2:]])


print plot(newpspec)
print plot(pspec)
```

```python
#flattening power spectrum to get peaks

slope_params = np.polyfit(f, newpspec, 2) # fit the curve of the PSD to isolate the oscillatory components
#np.polyfit gets vector components that minimize r^2 (polynomial function closest to all points)
slopefit = np.polyval(slope_params, f)
#fitting a 2nd degree polynomial to a graph of pspec using the paramaters defined by polyfit
# y = 5.14347266e-06 * x^2 -1.21294158e-02 * x + 6.21383500e-01

flatspec = newpspec-slopefit # subtract out the slope fit from the spectrum to flatten it out
#pretty neat

plt.plot(f, newpspec)
plt.plot(f, slopefit)
plt.plot(f, flatspec)
plt.legend(['original psd', 'PSD slope fit', 'flatspec (psd - 1/f)'], loc=1)
```

```python
#Finding peaks
peakind = sp.signal.find_peaks_cwt(flatspec, np.arange(0.1,10), wavelet=None)

peakind = np.array(peakind)
peakind = peakind[f[peakind]<=40]
#we only want the x values of peaks in the range we expect to see them i.e. the first 3 big ones
peakvals = flatspec[peakind]
#gets y vals of the peaks we found

plt.plot(f, flatspec)
plt.plot(f[peakind], flatspec[peakind], 'or')
plt.legend(['flatspec (psd - 1/f)', 'putative peaks'], loc=1)
```

```python
def norm(x, *args):
    number_of_gaussians = int(args[-1]) # number of gaussians from peak detection
    fitdat = np.array(0)
    for peaki in range(number_of_gaussians):
        fitdat = fitdat + args[peaki+(2*number_of_gaussians)]*sp.stats.norm.pdf(x, loc=args[peaki], scale=args[peaki+number_of_gaussians]) # add a bunch of gaussians together
    return fitdat
```

```python
#fit combination of gausians (norm function) to data
params = list(f[peakind]) + [2, 2, 4] + list(peakvals) + [len(peakind)] # initial guess for parameters based on peak detection
oscillatory_fit_params,_ = sp.optimize.curve_fit(norm, f, flatspec, p0=params) # fitting parameters
oscillatory_fit = norm(f, *oscillatory_fit_params) # getting gausian values that represent the peaks

plt.plot(f, flatspec)
#Flat power spectrum (because we subtracted the slopefit from each point)

norm(f, *oscillatory_fit_params)

#print f
plt.plot(f, oscillatory_fit)

plt.plot(f, pspec)
#vanilla power spectral density
plt.plot(f, pspec-oscillatory_fit)
#psd minus first 3 big oscillation peaks
plt.legend(['flatspec (psd - 1/f)','oscillatory fit', 'original psd', 'no oscillations'], loc=1)
```

```python
#plot fitted spectrum and original spectrum
plt.plot(f, pspec)
plt.plot(f, (slopefit + oscillatory_fit))

plt.legend(['original spectrum', 'fitted spectrum'])
```
